```python
def RootFileToDataFrame(fileName, branchList, onlyFirstElementOfLists=False):
    with uproot3.open(fileName) as rootFile:
        tree = rootFile['Lb_T/DecayTree']
        ## Load dataFrame with desired branches
        df = tree.pandas.df(branchList, flatten=False)

        ## Reset indices
        df.reset_index(drop=True, inplace=True)
        
        return df
```

```python
def SplitColumnOfTuples(df, column, columnSuffixes, dropOriginal=True, ignoreIfNonExisting=True):
    if ignoreIfNonExisting:
        if column not in list(df):
            return
    
    newColumns = [column + suffix for suffix in columnSuffixes]
    df[newColumns] = pd.DataFrame(df[column].tolist(), index=df.index)
    
    if dropOriginal:
        df.drop(column, axis='columns', inplace=True)
```

```python
def SplitColumnOfMatrices(df, column, dropOriginal=True, ignoreIfNonExisting=True):
    if ignoreIfNonExisting:
        if column not in list(df):
            return
    
    nMatrixRows = len(df[column][0])
    tempColumns = [column + '_' + str(row) +  '_' for row in range(nMatrixRows)]
    df[tempColumns] = pd.DataFrame(df[column].tolist(), index=df.index)
    nMatrixColumns = len(df[tempColumns[0]][0])
    
    for row, tempDFColumn in zip(range(nMatrixRows), tempColumns):
        tempDF = pd.DataFrame(df[tempDFColumn].tolist(), index=df.index)
        for col in range(nMatrixColumns):
            if col >= row:
                df[tempDFColumn + str(col)] = tempDF[col]
                
    df.drop(tempColumns, axis='columns', inplace=True)  
                
    if dropOriginal:
        df.drop(column, axis='columns', inplace=True)   
```

```python
for feature in ['seed_vtx', 'p_refPoint', 'p_momentum', 'pim_refPoint', 'pim_momentum']:
    SplitColumnOfTuples(df_log, feature, ['_X', '_Y', '_Z'], dropOriginal=True)
```

```python
for matrix in ['seed_ci', 'p_posMomCovMatrix', 'pim_posMomCovMatrix']:
    SplitColumnOfMatrices(df_log, matrix, dropOriginal=True)
```
