```python
def root_file_to_df(
    filename: str,
    branchlist: List[str]
) -> pd.DataFrame:
    with uproot3.open(filename) as file_root:
        tree = file_root['Lb_T/DecayTree']
        ## Load dataFrame with desired branches
        df = tree.pandas.df(branchlist, flatten=False)

        ## Reset indices
        df.reset_index(drop=True, inplace=True)
        
        return df
```

```python
def split_column_of_tuples(
    df: pd.DataFrame,
    column: str,
    column_suffixes: List[str],
    drop_original: bool = True,
    ignore_if_non_existing: bool = True
) -> None:
    if ignore_if_non_existing:
        if column not in list(df):
            return
    
    new_columns = [column + suffix for suffix in column_suffixes]
    df[new_columns] = pd.DataFrame(df[column].tolist(), index=df.index)
    
    if drop_original:
        df.drop(column, axis='columns', inplace=True)
```

```python
def split_column_of_matrices(
    df: pd.DataFrame,
    column: str,
    drop_original: bool = True,
    ignore_if_non_existing: bool = True
) -> None:
    if ignore_if_non_existing:
        if column not in list(df):
            return
    
    n_matrix_rows = len(df[column][0])
    temp_columns = [column + '_' + str(row) +  '_' for row in range(n_matrix_rows)]
    df[temp_columns] = pd.DataFrame(df[column].tolist(), index=df.index)
    n_matrix_columns = len(df[temp_columns[0]][0])
    
    for row, temp_df_column in zip(range(n_matrix_rows), temp_columns):
        temp_df = pd.DataFrame(df[temp_df_column].tolist(), index=df.index)
        for col in range(n_matrix_columns):
            if col >= row:
                df[temp_df_column + str(col)] = temp_df[col]
                
    df.drop(temp_columns, axis='columns', inplace=True)  
                
    if drop_original:
        df.drop(column, axis='columns', inplace=True)   
```

I have the feeling that some events with more than one element in a certain column are counted as separate events in the file_log (i.e. same pion, two protons). We need to split those rows in two in order to provide an efficient truth match.

There is a function that does this: `pandas.explode`; however, its key feature (multi-column explode) is only available in later versions of Pandas requiring Python >3.7. What follows is exactly what it looks like.

```python
from pandas.core.dtypes.common import is_scalar, is_list_like
from pandas.core.indexes import base as ibase

def explode(
    df,
    column,
    ignore_index=False,
):
    if not df.columns.is_unique:
        raise ValueError("columns must be unique")

    columns: list[str | tuple]
    if is_scalar(column) or isinstance(column, tuple):
        assert isinstance(column, (str, tuple))
        columns = [column]
    elif isinstance(column, list) and all(
        map(lambda c: is_scalar(c) or isinstance(c, tuple), column)
    ):
        if not column:
            raise ValueError("column must be nonempty")
        if len(column) > len(set(column)):
            raise ValueError("column must be unique")
        columns = column
    else:
        raise ValueError("column must be a scalar, tuple, or list thereof")

    df = df.reset_index(drop=True)
    if len(columns) == 1:
        result = df[columns[0]].explode()
    else:
        mylen = lambda x: len(x) if is_list_like(x) else -1
        counts0 = df[columns[0]].apply(mylen)
        for c in columns[1:]:
            if not all(counts0 == df[c].apply(mylen)):
                raise ValueError("columns must have matching element counts")
        result = pd.DataFrame({c: df[c].explode() for c in columns})
    result = df.drop(columns, axis=1).join(result)
    if ignore_index:
        result.index = ibase.default_index(len(result))
    else:
        result.index = df.index.take(result.index)
    result = result.reindex(columns=df.columns, copy=False)

    return result
```

```python
## Warning: this is very slow. Event per event, in fact. I don't think
## there's a better way to implement this. merge_columns need to be the same
## between the two DataFrames.
def truth_match(
    df_root:      pd.DataFrame,
    df_log:       pd.DataFrame,
    merge_columns: List[str]
) -> pd.DataFrame:
    
    root_columns = list(df_root)
    root_columns_to_keep = [column for column in root_columns if column not in merge_columns]
    
    list_of_matched_events = []
    
    ## List of indices for df_root. These are the indices of events to match
    root_indices = list(range(len(df_root)))
    ## List of indices for df_root. These are the indices of unpaired events
    log_indices = list(range(len(df_log)))
    
    for root_index in root_indices:
        ## Get the event
        evt_to_match = df_root.iloc[root_index]
        ## Compute the total sum of absolute discrepancies between the two DF on merge_columns
        ## Use relative discrepancies: otherwise a small discrepancy on a big value (say,
        ## momenta components in position-momentum covariance matrices) may offset many larger
        ## discrepancies on smaller values.
        root_log_discrepancies = abs(
            (evt_to_match[merge_columns] - df_log[merge_columns].iloc[log_indices])/evt_to_match[merge_columns]
        ).sum(axis=1)
        ## Get minimum discrepancy and corresponding index. We assume that's the matching event
        matched_index = root_log_discrepancies.idxmin()
        discrepancy = root_log_discrepancies.min()
        ## Remove this event to prevent double matches
        log_indices.remove(matched_index) ## You thought I would delete the row, didn't you? Gotcha.
        ## Create the row for future DF (as dictionary, we need all the time efficiency we can get)
        evt_root = evt_to_match[root_columns_to_keep]
        evt_log = df_log.iloc[matched_index]
        evt_dic = pd.concat([evt_root, evt_log]).to_dict()
        evt_dic['match_index'] = matched_index
        evt_dic['mismatch_perc'] = discrepancy
        list_of_matched_events.append(evt_dic)
    
    return pd.DataFrame(list_of_matched_events)
```
